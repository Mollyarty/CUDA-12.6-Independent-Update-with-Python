```python
import os
```

```python
files = [f for f in os.listdir("./")]
```

```python
files.pop(0)
files.pop()
files.pop()
print(files)
```

```python
folder_files = [f for f in os.listdir("./"+files[0]) if not f.endswith(('.md'))]
print(folder_files)
print(len(folder_files))
```

```python
master_list = []
total = 0
for folder in files:
    folder_files = [f for f in os.listdir("./"+folder) if not f.endswith(('.md'))]
    for subfolder in folder_files:
        #print(folder+"/"+subfolder)
        subfolder_files = [f for f in os.listdir("./"+folder+"/"+subfolder) if f.endswith(('.vcxproj'))]
        cur_len = len(subfolder_files)
        total += cur_len
        for sf in subfolder_files:
            master_list.append("./"+folder+"/"+subfolder+"/"+sf)

print(master_list)
```

```python
for fil in master_list:
    ofile = fil+""
    lines = []
    with open(ofile) as file:
        lines = file.readlines()
        #print(lines)
        for i in range(len(lines)):
            if "CUDA 12.5" in lines[i]:
                lines[i] = lines[i].replace("CUDA 12.5", "CUDA 12.6")
        file.close()
    output_path = "./testing" + fil[1:]
    output_split = output_path.split("/")
    output_path = ""
    for s in output_split[:-1]:
        output_path += s+"/"
    print(output_path)
    output_dir = os.path.dirname(output_path)
    os.makedirs(output_dir, exist_ok=True)
    with open("./testing" + fil[1:], 'w') as file:
        file.writelines(lines)
```
